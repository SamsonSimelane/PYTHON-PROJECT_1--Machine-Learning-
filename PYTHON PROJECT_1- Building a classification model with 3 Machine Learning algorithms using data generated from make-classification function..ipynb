# 1. Make synthetic dataset

## 1.1. Generate the dataset 

```python
from sklearn.datasets import make_classification 
X_Data,Y_Data=make_classification (n_samples = 50,n_classes =2, n_features=10, n_redundant =0, random_state=42)
```

## 1.2. Let's examine the data dimension 

We can see that there are 50 rows (samples) and 10 columns (features) for the X variable and 50 rows and 1 column (class label) for the Y variable. 

```python
X_Data.shape, Y_Data.shape 
```

```python
 # Print the X_Data dataset and renaming X_Data as df_X
df_X=pd.DataFrame(X_Data)
df_X
```

```python
 # Print the Y_Data dataset and renaming Y_Data as df_Y
df_Y=pd.DataFrame(Y_Data)
df_Y
```

```python
 # Representing the X_Data dataset and Y_Data dataset as a single data frame and naming the data frame final_dataset
final_dataset =pd.concat([df_X,df_Y], axis = 1) 
final_dataset
```

###  2)  Correlation Matrix.  

#### A correlation matrix is tool that allows the visualization of the strengths and weakness between features. 

#### The main diagonal gives the positive strong correlation.The correlation is understood to be between +1 and -1. Both +1 and -1 are intepreted to be representing strong correlation, where +1 is to the positive direction and -1 is to the negative direction. 

#### The desired correlation is the positive correlation between the features and the target.

```python
matrix=final_dataset.corr().round(3) # Computing the correlation and round off to 3 decimal places.
matrix
```

###  2.1)  Heatmap.  

```python
plt.figure(figsize=(10,10))
sns.heatmap(matrix, cmap ="RdBu", vmin=-1, vmax = 1 ,annot = True, center =0, square = True)
```

#### NOTE: On the heatmap representation each cell below the main diagonal is reflected about the main diagonal. Thus, the bottom segment below the main diagonal,  is identical to the top segment above the main diagonal. 

#### To simplified representation we hide the above repeated segment of the heatmap.

```python
mask =np.triu(np.ones_like(matrix, dtype = bool))
mask
```

#### NOTE: The best way to analyizing correlation matrix is to first unstack the correlation matrix as below:

```python
matrix = matrix.unstack()
matrix 
```

### 3. Data split (70/30 ratio) 

#### 3.1. Data split 

#### A ratio of 70/30 is used for data splitting such that 70% goes to the training subset and 30% goes to the testing subset.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_Data,Y_Data, test_size = 0.3, random_state=42)
```

### 4.) Building a Machine learning model using 3 algorithms: 1. Suppport vector machine (SVM), 2. Random Forest Classifier  3. Logistic regression. 

```python
from sklearn import svm
from sklearn.ensemble  import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    'svm': {
     'model': svm.SVC(gamma ='auto'),
      'params': {
          'C' : [1,10,200],
          'kernel':['rbf', 'linear']
      }
    },
   'random_forest':{
    'model' : RandomForestClassifier(),
    'params' :{ 
    'n_estimators':[1,5,100]
    }
        
  },
    'logistic_regression':{
    'model': LogisticRegression(solver = 'liblinear', multi_class ='auto'),
    'params': {
      'C' : [1,5,100]
      }
    }
}
```

### 5.) Hyperparameter Tuning 

####  Performing the hyperparameters tuning of all the three models  using the GridSearchCV. The GridSearchCV is used to search through hyper-parameters. 

```python
from sklearn.model_selection import GridSearchCV
```

```python
scores = []
for model_name, np in model_params.items():
    grid = GridSearchCV(np['model'], np['params'], cv =5, return_train_score =False)
    grid.fit(x_train, y_train)
    scores.append({
    'model': model_name,
    'best_score': grid.best_score_,
    'best_params':grid.best_params_
})
```

```python
df =pd.DataFrame(scores, columns =['model','best_score','best_params'])
df
```
